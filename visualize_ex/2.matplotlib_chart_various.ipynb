```python
# 데이터 프레임 생성
import pandas as pd
import numpy as np

# 날짜 범위 생성 (예시로 2021년 1월부터 2021년 12월까지)
# pd.date_range(start, end, freq)
# # - start: 생성할 날짜 범위의 시작일을 지정합니다. 문자열 형식인 경우 YYYY-MM-DD 형식을 따라야 합니다.
# # - end: 생성할 날짜 범위의 종료일을 지정합니다. 문자열 형식인 경우 YYYY-MM-DD 형식을 따라야 합니다.
# # - freq: 날짜 간격(frequency)을 지정
dates = pd.date_range(start='2021-01-01', end='2021-12-31', freq='M')

# 판매량 데이터 생성
# - np.random.normal(loc, scale, size) : 정규 분포에서 난수 생성
# # - loc: 평균(mean) 값으로, 생성된 난수들의 평균 지정
# # - scale: 표준 편차(standard deviation) 값으로, 생성된 난수들의 표준 편차를 지정
# # - size: 생성할 난수의 개수 지정, 생성된 배열의 크기를 결정하는데 사용
np.random.seed(0)
product_a_sales = np.random.normal(loc=100, scale=20, size=len(dates))
product_b_sales = np.random.normal(loc=150, scale=30, size=len(dates))
product_c_sales = np.random.normal(loc=80, scale=15, size=len(dates))

# 데이터프레임 생성
sales_data = pd.DataFrame({
    'Date': dates,
    'Product A': product_a_sales,
    'Product B': product_b_sales,
    'Product C': product_c_sales
})

# 데이터프레임 출력
print(sales_data.head())
```

```python
# 원 그래프(Pie Chart)
# 과일의 크기 비율 시각화
import matplotlib.pyplot as plt

# 데이터
fruits = ['Apple', 'Banana', 'Orange', 'Grape', 'Watermelon']
sales = [30, 40, 25, 15, 20]

# 파이차트 생성
# plt.pie(실제값,라벨s,퍼센트,시작위치)
plt.pie(sales, labels=fruits, autopct='%1.1f%%', startangle=90)

# 차트 제목 설정
plt.title('Sales of Fruits')

# 차트 보여주기
plt.show()
```

```python
# 도넛 그래프(Donut Chart)
# 삼품 카테고리별
import matplotlib.pyplot as plt

# 데이터
categories = ['Category A', 'Category B', 'Category C']
ratios = [30, 40, 20]

# 도넛 그래프 생성
# plt.pie(데이터, labels=레이블, autopct='%비율백분율로표시%', startangle=시작_각도, wedgeprops={'edgecolor': 'white'})
plt.pie(ratios, labels=categories, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor':'white'}, pctdistance=.75)

# 중앙에 흰색 원 생성하여 도넛 모양으로 만들기 
# plt.Circle((0, 0), 0.5, color='white') : 중앙점에서부터 반지름의 0.5 지점까지 흰색으로 채우기
# center_circle = plt.Circle((0, 0), 0.5, color='white', edgecolor='white')
center_circle = plt.Circle((0,0), 0.5, color='white')

# plt.gcf() 함수로 현재 그래프의 Figure 객체를 가져온 뒤 add_artist()로 원을 추가함
fig = plt.gcf()
fig.gca().add_artist(center_circle)

# 그래프 제목 설정
plt.title('Category Ratios')

# 도넛 그래프 보여주기
plt.show()
```

```python
# 히트맵(Heatmap):
# 데이터의 상대적인 크기를 색상으로 시각화하여 나타내는 그래프
# 주로 2차원 데이터를 표현하며, 행과 열로 구성된 격자 형태로 표시
# 데이터의 패턴, 상관관계, 분포 등을 시각적으로 파악할 수 있어 다양한 분야에서 사용됨

# fmt 매개변수 : 값을 표현하는 형식 설정
# # - g : 일반숫자
# # - .2f : 소수점 이하 2자리
# # - .0f : 소수점 이하 자리수 0 => 정수 형식
# # - d : 10진수로 표시
# # - '%':.2f : 백분율 형식으로 값 표시, 소수점 이하 2자리까지 표시
```

```python
# 데이터 준비
import numpy as np
import pandas as pd

# 데이터
data = {
    'Date': ['2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04', '2021-01-05'],
    'Stock1': [100, 110, 105, 95, 120],
    'Stock2': [80, 85, 90, 95, 100],
    'Stock3': [120, 130, 125, 135, 140],
    'Stock4': [70, 75, 80, 85, 90]
}

# 데이터프레임 생성
df = pd.DataFrame(data)
df
```

```python
# matplotlib으로 그리기
import matplotlib.pyplot as plt

# 데이터 인덱스 설정
df.set_index('Date', inplace=True)

# 히트맵 생성
fig, ax = plt.subplots()
im = ax.imshow(df.values, cmap='YlOrRd')

# 축 설정
ax.set_xticks(np.arange(len(df.columns)))
ax.set_yticks(np.arange(len(df.index)))
ax.set_xticklabels(df.columns)
ax.set_yticklabels(df.index)

# 텍스트 주석 추가
for i in range(len(df.index)):
    for j in range(len(df.columns)):
        text = ax.text(j, i, df.values[i, j], ha='center', va='center', color='black')

# 컬러바 추가
cbar = ax.figure.colorbar(im, ax=ax)

# 그래프 제목 설정
plt.title('Stock Prices Heatmap')

# 그래프 보여주기
plt.show()
```

```python
# seaborn으로 그리기
# pip install seaborn

import numpy as np
import pandas as pd
import seaborn as sns

# 데이터 인덱스 설정
df.set_index('Date', inplace=True)

# 히트맵 생성
# cmap 종류 : YlOrRd, viridis, plasma,inferno, magma, cividis, cool, coolwarm, hot, RdYlBu, BuPu
# fmt='d' 는 정수 형태(integer format)로 숫자를 입력하라는 뜻
sns.heatmap(df, cmap='viridis', annot=True, fmt='g')

# 그래프 제목 설정
plt.title('Stock Prices Heatmap')

# 그래프 보여주기
plt.show()
```

```python
cmap의 종류
sns.heatmap(df, cmap='YlOrRd', annot=True, fmt='g')

"viridis": 연속적인 밝기 변화를 가지는 그린과 블루 컬러의 색상 맵
"plasma": 연속적인 밝기 변화를 가지는 보라색과 빨간색 컬러의 색상 맵
"inferno": 연속적인 밝기 변화를 가지는 주로 오렌지와 검은색 컬러의 색상 맵
"magma": 연속적인 밝기 변화를 가지는 보라색과 노란색 컬러의 색상 맵
"cividis": 연속적인 밝기 변화를 가지는 옅은 황색과 짙은 검은색 컬러의 색상 맵
"cool": 차가운 톤의 컬러로 이루어진 색상 맵
"coolwarm": 차가운 색상과 따뜻한 색상을 혼합한 색상 맵
"hot": 따뜻한 톤의 컬러로 이루어진 색상 맵
"RdYlBu": 빨간색에서 노란색, 파란색으로 이루어진 컬러 맵
"BuPu": 파란색과 보라색을 혼합한 컬러 맵
```

```python
# 네트워크 그래프(Network Graph)
# pip install networkx
# 현실 세계에서의 다양한 상호 관계를 시각적으로 나타내는 그래프
# 네트워크 그래프는 노드(Node)와 엣지(Edge)로 이루어져 있으며, 노드는 개별 요소 또는 개체를 나타내고, 엣지는 노드 간의 관계를 나타냄
# 소셜 네트워크, 웹 페이지 간의 링크 구조, 전력 공급망, 교통 네트워크 등 다양한 영역에서 사용

import networkx as nx
import matplotlib.pyplot as plt

# 노드와 엣지 데이터 생성
nodes = ['A', 'B', 'C', 'D', 'E']
edges = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('E', 'A')]

# 그래프 생성
G = nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)

# 그래프 시각화
pos = nx.circular_layout(G)
nx.draw_networkx(G, pos, with_labels=True, node_color='lightblue', edge_color='gray')
plt.axis('off')
plt.show()
```

```python
# 트리맵(Treemap):
# pip install squarify
# 계층적인 데이터를 사각형의 크기와 색상으로 시각화하는 방법
# 주로 계층적인 구조를 가진 데이터를 표현하고 비교하는 데 사용
# 트리맵은 데이터의 상대적인 크기와 구조를 한눈에 파악할 수 있어 효과적인 정보 전달을 가능

import squarify
import matplotlib.pyplot as plt

# 데이터와 라벨 생성
data = [400, 300, 200, 100]
labels = ['A', 'B', 'C', 'D']

# 트리맵 생성
plt.figure(figsize=(8, 8))
squarify.plot(sizes=data, label=labels, color=['lightblue', 'lightgreen', 'lightyellow', 'lightpink'])
plt.axis('off')
plt.show()
```

```python
# 국도도(Map of Connections):
# 노드(Node)와 엣지(Edge)로 이루어져 있으며, 노드는 위치를 나타내고, 엣지는 위치 간의 연결 관계를 나타냄
# 주로 지리적으로 분포된 데이터의 상호 연결성을 시각화하는 데 사용함

import networkx as nx
import matplotlib.pyplot as plt

# 노드와 엣지 데이터 생성
nodes = ['A', 'B', 'C', 'D', 'E', 'F']
edges = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('E', 'A'),('E', 'F')]

# 그래프 생성
G = nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)

# 노드 위치 정보 생성
pos = {'A': (0, 0), 'B': (1, 1), 'C': (2, 0), 'D': (3, 1), 'E': (4, 0), 'F': (5, 1)}

# 그래프 시각화
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray')
plt.show()
```
