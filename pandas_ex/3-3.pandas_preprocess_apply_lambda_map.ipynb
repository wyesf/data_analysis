### 1-2. apply () - 정의한 함수 DataFrame에 일괄적용
남자: 1, 여자: 0, 그밖에: -1

```python
df = pd.read_csv('datas/korea-star.csv')
df
```

```python
# 남자: 1 여자: 0 기타: -1
# [주의] 반드시 return 값이 존재 해야함

# 먼저 apply에 활용할 함수를 정의합니다.
def male_or_female(x):
    if x == '남자':
        return 1
    elif x == '여자':
        return 0
    else:
        return -1
```

```python
# pandas 내장함수 apply
# df['성별'] = df['성별'].apply(male_or_female)
df['성별_NEW'] = df['성별'].apply(male_or_female)
```

```python
df.head()
```

```python
# 파이썬 내장함수 map() 사용
# 벡터값을 대입해서 값을 확인하므로 pandas 내장함수 apply() 와 유사
def a_func(z) :
    return z * 2

a = 3
print(a_func(a))

a_list = [1,2,3,4]

result = map(a_func, a_list)  #여기서 list로 싸주거나 아래에서 list 싸주면 값을 확인
# print(result)
print(list(result))
```

```python
# [문제] apply 함수 활용 및 컬럼 추가
(목표) cm당 인기지수 구하기
(인기지수/ 키)를 구하여 '인기지수/키' 컬럼 추가
키: 174, 인기지수: 100000
계산 식: 100000 / 174
```

```python
# 1. apply에 적용할할 함수 정의
def cm_to_brand(x) :
    value = x['인기지수'] / x['키']
    return round(value, 2)

# mine
# def cm_to_brand(x) :
#     y = x['인기지수'] / x['키']
#     # print(y)
#     return
#     # return y  # 얘를 안해줌
# df['인기지수/키'] = df.apply(cm_to_brand, axis=1)
# df.head()

# 동석님
# def cm_to_brand(row) :
#     height = row['키']
#     pop = row['인기지수']
#     result = pop / height
#     return result
# df['인기지수/키'] = round(df.apply(cm_to_brand, axis=1),2)
# df.head()

# 함수 사용 안할 경우
# df['인기지수/키' ]= df['인기지수']/df['키']
# df
```

```python
# 방법 1 : 데이터프레임을 전부 사용
df['인기지수/키'] = df.apply(cm_to_brand, axis=1)
# df['인기지수/키'] = df.apply(cm_to_brand) 는 오류남 디폴트가 axis=0 방향이므로
# 함수가 한번씩 실행될 때마다 두개의 컬럼을 참조해야 하기 때문에 axis=1을 명시해야 함
df.head()
```

```python
# 방법 2 : 해당 시리즈만 사용(인기지수/키)
df['인기지수/키'] = df[['인기지수','키']].apply(cm_to_brand, axis=1)
df.head()
```

### 1-3. lambda 함수의 적용
- lambda는 1줄로 작성하는 간단 함수식
- return을 별도로 명기하지 않음

```python
# lambda 함수 : 한줄 함수
# 식별자(변수) = lambda 매개변수1, 매개변수2 : 수행문
# f = lambda x, y : x + y
# 실행결과_식별 = f(5,6)

f = lambda x, y : x + y
f(5,6)
```

```python
def aaa(l,m) :
    return l * m

aaa(f(5,6), 2)  #전달인자를 넣어줌
```

### lambda 함수를 직접 함수에 사용하기

```python
# 오류남
# def aaa(l,m) :
#     return l * m
# x = 5
# y = 6

# aaa(lambda x, y : x + y, 2)
```

```python
# 파이썬의 내장함수 map() 함수와 lambda 함수 사용 예시
c = [1,2,3,4]
map(lambda x : x * 2, c)
```

```python
df['성별'].apply(f)
```

```python
# lambda 함수 :간단한 계산식을 적용할 때 많이 사용
```

- 간단한 계산식은 lambda로 적용

### 3. map() - 값을 매핑!

```python
# pandas DataFrame의 map() 메서드 사용
```

```python
df = pd.read_csv('datas/korea-star.csv')
df.head()
```

#### df.map(mapping) 사용 방법
- mapping : dictionary(딕셔너리) 자료구조 값이어야 함
<pre>
{
    원래값1 : 변경값1,
    원래값2 : 변경값2,
    원래값3 : 변경값3
}
</pre>

```python
# 값 counts 확인
df['성별'].value_counts()
```

```python
mapping = {
    '여자' : 1,
    '남자' : 0
}
```

```python
# 여기로 전달!
df['성별'] = df['성별'].map(mapping)
```

```python
df
```

```python
df.info()
```

```python
# 데이터 타입이 맞아야 함 (특히 숫자인 경우)
# mapping = {
#     '1' : 'F',
#     '0' : 'M'
# }
# 원래값을 지정할 때 dtype을 맞춰야 함
mapping = {
    1 : 'F',
    0 : 'M'
}
```

```python
# 여기로 전달!
df['성별'] = df['성별'].map(mapping)
df
```
