```python
# 2. 원핫인코딩 (One-hot-encoding)
# 원핫인코딩은 한개의 요소는 True 그리고 나머지 요소는 False로 만들어 주는 기법
# 원핫인코딩이 왜 필요할까요?
```

```python
df = pd.read_csv('datas/korea-star.csv')
df.head()
```

```python
blood_map = {
    'A': 0, 
    'B': 1,
    'AB': 2, 
    'O': 3,
}
```

```python
df['혈액형_code'] = df['혈액형'].map(blood_map)
```

```python
df.head()
```

```python
df['혈액형_code'].value_counts()
```

```python
우리가 만약 df['혈액형_code']를 머신러닝 알고리즘에 그대로 넣어 데이터를 예측고자 한다면,
컴퓨터는 '혈액형_code'안에서 값들간의 관계를 스스로 형성하게 된다.

이 상황에서 만약 B형은 1, AB형은 2라는 값을 가지고 있는데,
컴퓨터는 B형 + AB형 = O형이다라고 잘못 관계를 맺을 수 있게 된다.
따라서, 서로 개별적으로 인식 할수 있도록 column을 형성해 주는 것이 필요하다.
해당 column에는 True 나머지는 모두 False를 넣어 준다.
A, B, AB, O형의 관계는 독립적이다는 표현을 한다. *이를 원핫인코딩이라고 한다.
```

```python
df['혈액형_code']
```

```python
# one_hot_encoding할 컬럼 지정

# get_dummies()
# 범주형(문자형) 데이터를 one-hot-encoding으로 바꿔주는 함수
# pd.get_dummies(데이터프레임, prefix='접두어', dtype=데이터타입)

# pd.get_dummies(df['혈액형_code'])
# pd.get_dummies(df['혈액형_code'], prefix='혈액형_', dtype='int')

# 데이터 프레임을 넣어도 됨
pd.get_dummies(df)
# pd.get_dummies(df, dtype='int')
```

```python
# prefix를 설정하려면?
```

```python
# 컬럼이름 앞쪽에 붙이기
pd.get_dummies(df['혈액형_code'], prefix='blood_type')
```

```python
# 숫자로 표현
pd.get_dummies(df['혈액형_code'], prefix='blood_type', dtype=int)
```

```python
# 문제 : 복수개의 컬럼을 동시에 one-hot-encoding 적용
# 소속사, 성별, 혈액형
# pd.get_dummies(df[['소속사','성별','혈액형']], dtype=int)
pd.get_dummies(df, columns=['소속사','성별','혈액형'], dtype=int)
```
