```python
# 문제
# 구글드라이브에 있는 "house_price.csv" 파일을 pandas DataFrame으로 loading 하기
# 로딩한 DataFrame의 다음 정보 알아보기
# # 컬럼명, 데이터갯수, 데이터 타입, 결측치 갯수를 알아보는 명령 수행하기
# 로딩한 DataFrame 의 기술 통계 정보 확인하기
# 컬럼명 바꾸기 : 분양가격(㎡) -> 분양가격
# 두 개의 키(분양가, 연도)로 내림차순 소팅 하기
# # 분양가가 가장 높았던 연도 확인
```

```python
hp = pd.read_csv("datas/house_price.csv")
hp
```

```python
hp.info()
```

```python
hp.describe()
```

```python
hp.columns
```

```python
new_colm = ['지역명', '규모구분', '연도', '월', '분양가격']
hp.columns = new_colm
hp.columns
```

```python
hp.head()
```

```python
hp.rename(columns={'분양가격':'분양가격(㎡)'},inplace=True)
hp.head()
```

```python
hp.sort_values(by=['분양가격(㎡)','연도'])
hp
```
